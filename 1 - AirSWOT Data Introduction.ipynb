## AirSWOT Overview

AirSWOT consists of three main instruments. The primary instrument is called the Ka-band SWOT Phenomenology Airborne Radar, or KaSPAR. KaSPAR is an interferometric synthetic aperture radar used by AirSWOT to measure surface water levels. The other two instruments are an Applanix POSAV 610 integrated GNSS and IMU (global navigation satellite system and inertial measurement unit), and a Cirrus color infrared digital camera system. The GNSS and IMU are used to precisely track the aircraft position and remove the effects of aircraft motion and attitude on the radar data. The digital camera can be used to map the extents of surface water in cloud-free conditions (however, in the Delta-X campaign, the camera was not installed as AVIRIS-NG was collecting data).

AirSWOT uses near-nadir wide-swath radar interferometry (Rodriguez et al., 2017; Altenau et al., 2017) to measure water-surface elevation. The instrument operates at Ka-band (35.75 GHz), allowing for a compact antenna configuration using a ~1×1-m plate. The instrument includes six antennas that can form multiple baseline pairs for along-track and across-track interferometry. These antenna combinations provide contiguous coverage from about 500 m to 4 km from nadir, removing water-motion-induced geolocation and height errors. The instrument is coupled with a high-precision Applanix GPS/IMU system capable of high-precision trajectory reconstruction. The GPS/IMU data is used to compensate for the effects of aircraft motion and attitude of the radar data. 

AirSWOT was developed in collaboration with the SWOT project and has supported multiple ocean and hydrology campaigns. In addition to the data from the pre-Delta-X campaign (Denbina et al., 2019), its performance has been demonstrated over the Tanana River (Altenau et al., 2017), the Yukon Flats Basin (Pitcher et al., 2019), the Willamette River (Tuozzolo et al., 2019), and Arctic-Boreal regions of Canada and Alaska (Fayne et al., 2019).

AirSWOT produces continuous gridded elevation data. On a pixel basis, elevation error is dominated by random noise, and contiguous pixels along a river channel or within a contiguous waterbody must be averaged to reduce measurement noise. Assuming a total averaging area of 1 km^2, the height noise is less than 10 cm. In addition to random noise, other factors such as atmospheric delays, GPS errors, and instrument drift produce slowly varying errors that are corrected using in situ data (leveled pressure transducers, for Delta-X) and cross-over calibration from overlapping flight lines. The combined measurement system yields errors smaller than 1 cm/km over 10-km reaches, which is close to a factor of 2 better than required for Delta-X.

AirSWOT is designed for integration into the standard Avcon port in a KingAir B200 and has flown in the past on the NASA AFRC KA B200 (#801). For Delta-X, it is hosted on a Dynamic Aviation King Air. There are three components to the overall instrument suite: 1) radar mounted in the nadir forward port of the aircraft; 2) Avcon electronics rack in the main cabin; and 3) Welch power distribution rack in the rear of the aircraft. Integration requires two days. Two instrument operators are required in flight. Radiometric calibration data is acquired prior to deployment by imaging a corner reflector array in Rosamond, CA.

The L2 data products are GeoTIFF raster files in the Universal Transverse Mercator (UTM) coordinate system.  For the Delta-X domain, these products are in UTM Zone 15 North.  These files can be drag-and-dropped into QGIS for easy viewing.  For example:

If you look through the L2 files, you will notice that the .tif files in the L2 products have filenames that look like this:

**utm_m0m_20210401152540.hgt.tif**

But what do these filenames mean?

- "utm" denotes that this product is in the UTM coordinate system.
- "m0m" denotes the AirSWOT processing mode used to generate the product, which is "*m*ode *0* with *m*otion compensation".  As a science user, you do not need to worry about this, as "m0m" is the standard mode and, so far, all released AirSWOT data products are from this mode.
- "20210401152540" is a date and time tag denoting when the AirSWOT data was acquired, in the UTC timezone.  The date is formatted as YYYYMMDD, followed by the time formatted as HHMMSS.  For this example, the data was acquired at approximately April 1, 2021 at 15:25:40 UTC time.  Note that each AirSWOT flight line, since it was collected by an aircraft-mounted instrument, can span multiple minutes of flight time.  The time tags in the filename denote the approximate start time of the data.
- "hgt" denotes that the file contains the heights measured by AirSWOT for each image pixel.  For L2 data products, these heights are referenced to the WGS84 ellipsoid.  We will discuss more about vertical datums, and converting the AirSWOT data to other datums such as the North American Vertical Datum (NAVD88), in the next module.  Note that there are other data products besides the "hgt" files, but we will discuss them later in this section.
- ".tif" denotes that this is a GeoTIFF file.  The L2 data includes plain-text metadata, but the raster data itself is all in the GeoTIFF format.

These products can also be loaded into Python using the Geospatial Data Abstraction Library (GDAL).  In the following example, we will load AirSWOT L2 height and magnitude products using Python and GDAL, then display the data.  Note that you might need to change the "data_path" variable and filenames in the code below to point to the correct folder where you have downloaded or saved the AirSWOT L2 data products.

```python
# Setup Google Colab:
# For access to Google Drive files
from google.colab import files
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# For saving to Google Drive
from google.colab import drive
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
Gdrive = GoogleDrive(gauth)

drive.mount('/content/drive',force_remount=True)

# Path containing the data files.
data_path = '/content/drive/MyDrive/DeltaX_Workshop_2022/airswot_sample_data/'
```

```python
import os
import numpy as np
from osgeo import gdal
import matplotlib.pyplot as plt

# Filename of the GeoTIFF we wish to load, relative to data_path.
hgt_raster_file = 'L2/utm_m0m_20210401152540.hgt.tif'

# Load the GeoTIFF using GDAL.
hgt_product = gdal.Open(os.path.join(data_path, hgt_raster_file))
hgt_product_geo = hgt_product.GetGeoTransform()

# Read the data from the GeoTIFF into a NumPy array.
hgt_product_data = hgt_product.ReadAsArray()
hgt_product_data[hgt_product_data == -10000] = np.nan # Values of -10000 represent void data.

# What are the UTM X and Y coordinates for the extents of this image?
x_left = hgt_product_geo[0]
x_right = hgt_product_geo[0] + (hgt_product_data.shape[1] - 1) * hgt_product_geo[1]
y_top = hgt_product_geo[3]
y_bottom = hgt_product_geo[3] + (hgt_product_data.shape[0] - 1) * hgt_product_geo[5]
utm_extents = (x_left/1e3, x_right/1e3, y_bottom/1e3, y_top/1e3) # convert from meters to kilometers

# Let's display the data using matplotlib.
plt.figure(figsize=(8,12))
plt.imshow(hgt_product_data, vmin=-27, vmax=-24, extent=utm_extents)
plt.xlabel('UTM X (km)')
plt.ylabel('UTM Y (km)')
cb = plt.colorbar(label='AirSWOT Height Above WGS84 Ellipsoid (m)', shrink=0.5)
```

The above image shows the height measured by AirSWOT with respect to the WGS84 ellipsoid for an area including the Wax Lake Outlet.  The Wax Lake Outlet is a river channel stretching from Wax Lake near the top of the image to the Wax Lake Delta near the bottom of the image.

Don't worry if this is not clear.  Since the raster contains the vertical height of each pixel, the features in the scene are not as obvious as in an actual image of the scene.  Note that if you view the files in QGIS, you can overlay them on other data such as Google Satellite imagery which can help you locate features in the image.

**Note that the AirSWOT height data is not a map of just water surface elevation!**

The L2 data products contain an estimate of the surface height for each image pixel regardless of whether that pixel contains water, land, or wetland vegetation.

AirSWOT can only be used to accurately estimate water surface elevation (WSE) and water surface slope (WSS) for open water, unobstructed by vegetation.  In areas containing inundated vegetation, AirSWOT will receive radar backscatter from *both* the water surface and the vegetation.  In this case, the measured height will lie somewhere between the water surface elevation and the elevation of the top of the vegetation.  That is, we would expect the AirSWOT height estimate to be biased above the true water surface level.

In addition, data from AirSWOT, like all radar remote sensing, is noisier than you might expect if you traditionally use optical or multi-spectral remote sensing.

Therefore, in order to produce accurate water surface elevations from AirSWOT data, we need to use spatial averaging to beat down the noise.  We can calculate the mean height over a large open water area with a high degree of accuracy, but the height estimates at the level of individual pixels will be too noisy to use directly if we wish to obtain an accurate water level estimate.

We will discuss how to estimate accurate water levels from AirSWOT data in the next module.  For now, let's look at the other L2 data products besides the "hgt" product.  They are:

- "mag" denotes that the file contains the Ka-band radar backscatter **mag**nitude for each pixel, or put simply, the radar reflectivity.  Areas that reflect more Ka-band backscatter will have higher values in the data, while areas that reflect less Ka-band backscatter will have lower values.  This file contains the backscattered magnitude in linear units of power, which can be converted to a dB-scaled image by calculating 20*log10(mag).
- "mag_ns" denotes that the file contains the Ka-band radar backscatter magnitude for each pixel, after noise subtraction has been performed.  Radar imagery contains additive noise due to the thermal noise in the radar electronics, among other potential causes.  This file contains backscatter that has been corrected for an estimated noise level in the imagery.  Compared to the "mag" image, this file will have lower values in dark areas of the image where the radar signal was dominated by noise.  In the brighter parts of the image, there should be only a small difference between the "mag" and the "mag_ns" images.
- "err" denotes that the file contains an estimate of the AirSWOT height **err**or for each pixel.  These values are the theoretical 1-sigma standard deviation of the AirSWOT height, in meters.
- "corr" denotes that the file contains the radar **corr**elation for each pixel.  To measure height, AirSWOT actually collects multiple radar images at the same time, using antenna with slightly different spatial positions.  This correlation is the magnitude of the complex correlation calculated between the two radar images used to estimate the height in the "hgt" raster.  The value of the correlation data varies from 0 to 1.  The higher the value, the better, as higher values represent less noisy pixels.
- "inc" files contain the radar incidence angle for each pixel.  Smaller values denote that the radar viewing angle was closer to nadir (looking straight down from the aircraft), while larger values denote that the radar was looking more to the side.  These values are in units of radians.
- "dhdphi" files contain a measure of the sensitivity between vertical height and interferometric phase for each pixel, in units of meters per radian.  This file will be useful mainly to users with a lot of radar remote sensing experience, so if this description doesn't make sense, feel free to ignore this file.

Note that each flight line (e.g., "20210401152540" in the example above) will have files for "hgt", "mag", "mag_ns", "err", "corr", "inc", and "dhdphi".

For each of the above product types, we will now show an example of loading and displaying the product in Python, starting with the backscatter magnitude file.

```python
# Load the "mag" file using GDAL, and set void data to NaN (not-a-number).
mag_raster_file = 'L2/utm_m0m_20210401152540.mag.tif'
mag_product = gdal.Open(os.path.join(data_path, mag_raster_file))
mag_product_data = mag_product.ReadAsArray()
mag_product_data[mag_product_data == -10000] = np.nan

# Since this data represents a ratio of power...
# It is more appropriately displayed on a log scale, rather than a linear scale:
mag_db = 20*np.log10(mag_product_data)

plt.figure(figsize=(8,12))
plt.imshow(mag_db, vmin=-5, vmax=15, cmap='gray')
plt.axis('off')
cb = plt.colorbar(label='AirSWOT Backscatter (dB)', shrink=0.5)
```

Note that the image data is oriented in the direction of the AirSWOT flight line.  In the above image, the radar was looking towards the NW, so that the right side of the image is closest to the radar, and the left side of the image is farther from the radar.  Due to the geometry of the radar backscatter, this means that the areas closer to the radar (imaged at a steeper incidence angle) tend to be brighter, while the areas farther from the radar tend to be darker.

In addition, one can observe variation in backscatter due to different land types.  For example, in the above image, the water is generally brighter than the land, particularly for parts of the image closer to the radar.

Similar to the "mag" file above, let's take a look at the "mag_ns" file.

```python
# Load the file using GDAL, and set void data to NaN (not-a-number).
magns_raster_file = 'L2/utm_m0m_20210401152540.mag_ns.tif'
magns_product = gdal.Open(os.path.join(data_path, magns_raster_file))
magns_product_data = magns_product.ReadAsArray()
magns_product_data[magns_product_data == -10000] = np.nan

# Since this data represents a ratio of power...
# It is more appropriately displayed on a log scale, rather than a linear scale:
magns_db = 20*np.log10(magns_product_data)

plt.figure(figsize=(8,12))
plt.imshow(magns_db, vmin=-5, vmax=15, cmap='gray')
plt.axis('off')
cb = plt.colorbar(label='AirSWOT Backscatter w/ Noise Subtraction (dB)', shrink=0.5)
```

The "mag_ns" file, above, looks similar to the "mag" file, though with some subtle differences.

If you take the difference between the magnitude before and after noise subtraction, we can see that the magnitude decreases the most in the parts of the image that were already dark (i.e., where the received signal was low compared to the noise).

```python
plt.figure(figsize=(8,12))
plt.imshow(mag_db - magns_db, vmin=0, vmax=3, cmap='viridis')
plt.axis('off')
cb = plt.colorbar(label='AirSWOT Noise Correction Magnitude (dB)', shrink=0.5)
```

Moving on, let's take a look at the "err" product containing the estimated height error of each pixel.

```python
# Load the "err" file using GDAL, and set void data to NaN (not-a-number).
err_raster_file = 'L2/utm_m0m_20210401152540.err.tif'
err_product = gdal.Open(os.path.join(data_path, err_raster_file))
err_product_data = err_product.ReadAsArray()
err_product_data[err_product_data == -10000] = np.nan

plt.figure(figsize=(8,12))
plt.imshow(err_product_data, vmin=0, vmax=3, cmap='viridis')
plt.axis('off')
cb = plt.colorbar(label='AirSWOT Height Error $(\sigma)$ (m)', shrink=0.5)
```

In the parts of the swath far from the radar, the expected AirSWOT height error increases.  Indeed, the height error at the individual pixel level can be on the order of a few meters, which is a much larger error than is desirable, considering the small variation in topography throughout the Mississippi Delta.

In the next module, we will show how by masking, filtering, and averaging the AirSWOT data, we can achieve height errors with root mean square error on the order of 10 cm.

```python
# Load the "corr" file using GDAL, and set void data to NaN (not-a-number).
corr_raster_file = 'L2/utm_m0m_20210401152540.corr.tif'
corr_product = gdal.Open(os.path.join(data_path, corr_raster_file))
corr_product_data = corr_product.ReadAsArray()
corr_product_data[corr_product_data == -10000] = np.nan

plt.figure(figsize=(8,12))
plt.imshow(corr_product_data, vmin=0, vmax=1, cmap='viridis')
plt.axis('off')
cb = plt.colorbar(label='AirSWOT Correlation', shrink=0.5)
```

The above image shows the AirSWOT correlation.  In the areas with higher correlation, we also expect lower height errors.  Note that we generally see higher correlation over water than land.  This is the case for water which reflects more radar backscatter than the land (see the above "mag" and "mag_ns" images, for example).

```python
# Load the "inc" file using GDAL, and set void data to NaN (not-a-number).
inc_raster_file = 'L2/utm_m0m_20210401152540.inc.tif'
inc_product = gdal.Open(os.path.join(data_path, inc_raster_file))
inc_product_data = inc_product.ReadAsArray()
inc_product_data[inc_product_data == -10000] = np.nan

# Convert the incidence angles from radians to degrees before displaying.
inc_product_data = np.degrees(inc_product_data)

plt.figure(figsize=(8,12))
plt.imshow(inc_product_data, cmap='viridis')
plt.axis('off')
cb = plt.colorbar(label='AirSWOT Incidence Angle (degrees)', shrink=0.5)
```

In the above image, note that the areas closest to the radar have lower incidence angles, while the areas farther from the radar have higher incidence angles.  For these data products, the AirSWOT incidence angle ranges from around 3 degrees at the near edge of the data, up to 28 degrees at the far edge of the data.

Often, when analyzing AirSWOT data, we will filter out data near the edges of the image, particularly the parts of the image at high incidence angle where the height errors tend to be the largest.

Lastly, we will look at the "dhdphi" product:

```python
# Load the "dhdphi" file using GDAL, and set void data to NaN (not-a-number).
dhdphi_raster_file = 'L2/utm_m0m_20210401152540.dhdphi.tif'
dhdphi_product = gdal.Open(os.path.join(data_path, dhdphi_raster_file))
dhdphi_product_data = dhdphi_product.ReadAsArray()
dhdphi_product_data[dhdphi_product_data == -10000] = np.nan


plt.figure(figsize=(8,12))
plt.imshow(dhdphi_product_data, cmap='viridis')
plt.axis('off')
cb = plt.colorbar(label='AirSWOT dh/dphi (m/rad)', shrink=0.5)
```

The above image shows the conversion factor between the interferometric phase measured by the radar (in units of radians), and the corresponding vertical height (in meters).

This is the highest processing level of data, and L3 data products are not yet available for the Delta-X campaign (but they will be soon).  In this tutorial, we will use L3 data from the Pre-Delta-X campaign originally collected in 2015.

Each point contains an AirSWOT water surface elevation estimate. In the above image, we colored each point according to its "water_surface_elevation_NAVD88" value.  As we would expect, the higher elevation values are towards the top of the channel, farther from the ocean.

Now that we've demonstrated how to load the L3 data products into QGIS, let's go over a quick code example of how to load them into Python using Pandas.

(If you are unfamiliar with the Pandas library, think of a Pandas DataFrame as being similar to an Excel spreadsheet.)

```python
import pandas as pd

airswot = pd.read_csv(os.path.join(data_path, 'L3/PreDeltaX_WaterLevel_20150509_Atcha_AirSWOT.csv'))

print(airswot)
```

Using Pandas, the CSV data can be loaded into Python for further processing or analysis.  We will now go over the definition of each of the columns in the AirSWOT L3 CSV file.

- "channel_id" is a short string identifying the river channel that contains this data point.  For the Pre-Delta-X dataset, we only produced L3 data over the Wax Lake Outlet, identified as "WLO" in the data.  Future L3 data products will contain a greater number of river channels, as well as some lake data.
- "date" is the date the data was collected, formatted as YYYY-MM-DD.
- "time" is the UTC time the data was collected, using a 24-hour clock, formatted as HH:MM:SS.
- "utc_time" is the UTC time formatted as seconds of the day, rather than in the HH:MM:SS format.
- "along_channel_distance" is the distance along the river channel.  To calculate this, the AirSWOT data is transformed from geographic coordinates to an along-channel and cross-channel coordinate system defined by the river channel center line.
- "CoordX" is the X coordinate of the data point, in the UTM coordinate system.
- "CoordY" is the Y coordinate of the data point, in the UTM coordinate system.
- "longitude" is the longitude of the data point, in the WGS84 coordinate system, in units of degrees.
- "latitude" is the latitude of the data point, in the WGS84 coordinate system, in units of degrees.
- "water_surface_elevation_NAVD88" is the water surface elevation of the data point, in meters above the NAVD88 GEOID12B geoid model.
- "water_surface_elevation_WGS84" is the water surface elevation of the data point, in meters above the WGS84 ellipsoid.
- "water_surface_elevation_uncertainty" is the estimated uncertainty (1-sigma standard deviation) in the AirSWOT water surface elevation, in meters.  Note that these uncertainties are much less than the values in the L2 "err" product, as the heights have been filtered and averaged to produce the L3 product.

Now that we've loaded the data in Python, let's try plotting a water surface elevation profile along the Wax Lake Outlet.

```python
along = airswot['along_channel_distance'] / 1e3 # convert to km
wse = airswot['water_surface_elevation_NAVD88']

plt.figure(figsize=(10, 7))
plt.plot(along, wse, '-')
xlbl = plt.xlabel('WLO Along-Channel Distance (km)')
ylbl = plt.ylabel('AirSWOT Water Surface Elevation (m, NAVD88)')
```

This channel profile begins upstream near the Wax Lake itself, and ends at the Wax Lake Delta, close to the ocean.  From this profile, we can observe the slope of the water surface as the water flows through the Wax Lake Outlet and into the ocean.

In addition to the water surface elevation, it can also be useful to calculate the water surface slope.  Let's do that now.

```python
# Calculate slope, converting the water surface elevation into cm,
# so that the resulting slope has units of cm/km:
wss = np.gradient(wse * 100) / np.gradient(along)

# Plot slope vs. distance along Wax Lake Outlet:
plt.figure(figsize=(10, 7))
plt.plot(along, wss, '-')
xlbl = plt.xlabel('WLO Along-Channel Distance (km)')
ylbl = plt.ylabel('AirSWOT Water Surface Slope (cm/km)')

print('Mean Slope in Channel: {:.2f} cm/km.'.format(np.nanmean(wss)))
```

Unfortunately, this estimate of the slope is noisy and quantized, because it is calculated from profile data that is coarsely sampled in along-channel distance with a spacing of 50 m.

Instead of calculating the slope directly from the profile data, another approach is to filter the data using a [Savitzky-Golay](https://en.wikipedia.org/wiki/Savitzky–Golay_filter) filter, which will both smooth the data and calculate the derivative in one step.

```python
# The Savitzky-Golay filter is implemented in SciPy:
from scipy.signal import savgol_filter

# Filter the data...
# The window length of the filter needs to be in number of samples, not in distance.
# Let's try two filtering window sizes: 2 km and 10 km.
window_length_2km = int(2000 / 50) # Use a smoothing window of 2 km.
window_length_10km = int(10000 / 50) # Use a smoothing window of 10 km.
print('Window Length for 1 km Window: {}'.format(window_length_1km))
print('Window Length for 10 km Window: {}'.format(window_length_10km))
wss_filtered_2km = savgol_filter(wse * 100, window_length_2km, 2, deriv=1, delta=50/1e3, mode='nearest')
wss_filtered_10km = savgol_filter(wse * 100, window_length_10km, 2, deriv=1, delta=50/1e3, mode='nearest')

# Plot the original and filtered water surface slopes:
plt.figure(figsize=(10, 7))
plt.plot(along, wss, '-', label='Original')
plt.plot(along, wss_filtered_2km, '-', label='Filtered (2 km Window)')
plt.plot(along, wss_filtered_10km, '-', label='Filtered (10 km Window)')
xlbl = plt.xlabel('WLO Along-Channel Distance (km)')
ylbl = plt.ylabel('AirSWOT Water Surface Slope (cm/km)')
legend = plt.legend(loc=0, frameon=True, fontsize=8)
```

The slope estimates are now smoother.  We see that the water surface slope reaches its steepest point near the upstream part of the channel, around 5 km from the beginning of the profile.  After the water passes the Intracoastal Waterway the slope becomes more moderate, eventually settling around -2.5 cm/km.
